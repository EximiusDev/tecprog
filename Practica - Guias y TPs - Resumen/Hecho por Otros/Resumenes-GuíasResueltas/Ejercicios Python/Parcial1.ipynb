```python
from datetime import datetime
class Persona:
    def __init__(self,nombre:str,edad:int,dni:str):
        self.nombre=nombre
        self.edad=edad
        self.dni=dni
    def get_nombre(self):
        return self.nombre
    def get_edad(self):
        return self.edad
    def get_dni(self):
        return self.dni

class Legislador(Persona):
    def __init__(self,nombre:str,edad:int,dni:str):
        super().__init__(nombre,edad,dni)
        self.l_proyectos=[]
        self.l_presentaciones=[]
        self.l_asesores=[]
    def agregar_proyecto(self,proyecto):
        self.l_proyectos.append(proyecto)
    def presentar_proyecto(self,proyecto,fecha:datetime):
        self.l_presentaciones.append(PresentacionProyectoLey(fecha,proyecto,self))
    def agregar_asesores(self,asesor):
        self.l_asesores.append(asesor)
    def ver_presentaciones(self):
        for presentacion in self.l_presentaciones:
            presentacion.ver_presentacion()

class Diputado(Legislador):
    def __init__(self,nombre:str,edad:int,dni:str,partido:str):
        super().__init__(nombre,edad,dni)
        self.partido=partido
    def get_partido(self):
        return self.partido 
    
class Senador(Legislador):
    def __init__(self,nombre:str,edad:int,dni:str,provincia:str):
        super().__init__(nombre,edad,dni)
        self.provincia=provincia
    def get_provincia(self):
        return self.provincia
    
class ProyectoDeLey:
    def __init__(self,titulo:str,contenido:str):
        self.titulo=titulo
        self.contenido=contenido
        self.l_legisladores=[]
    def get_titulo(self):
        return self.titulo
    def get_contenido(self):
        return self.contenido
    def agregar_legislador(self,legislador):
        self.l_legisladores.append(legislador)

class PresentacionProyectoLey:
    def __init__(self,fecha:datetime,proyecto:ProyectoDeLey,legislador:Legislador):
        self.fecha=fecha
        self.proyecto=proyecto
        self.legislador=legislador
    def get_fecha(self):
        return self.fecha
    def ver_presentacion(self):
        print(f"En la fecha {self.fecha} el legislador {self.legislador.get_nombre()} presento el siguiente proyecto de ley:")
        print(self.proyecto.get_titulo())
        print(self.proyecto.get_contenido())
    
class Asesor(Persona):
    def __init__(self,nombre:str,edad:int,dni:str,especialidad:str):
        super().__init__(nombre,edad,dni)
        self.especialidad=especialidad
        self.l_legisladores=[]
    def get_especialidad(self):
        return self.especialidad
    def agregar_legislador(self,legislador):
        self.l_legisladores.append(legislador)

class Congreso:
    def __init__(self):
        self.l_personas=[]
    def agregar_persona(self,persona:Persona):
        self.l_personas.append(persona)
    def ver_presentaciones(self):
        for persona in self.l_personas:
            if(isinstance(persona,Asesor)==False):
                persona.ver_presentaciones()
                
if __name__=="__main__":
    diputado1=Diputado("Carlos Ferreira",43,"25032432","Juntos por el cambio")
    diputado2=Diputado("Miguel Fernandez",57,"18022432","Frente para la victoria")
    senador1=Senador("Marcos Solis",36,"35432170","Santa Fe")
    senador2=Senador("Alejandro Farias",39,"31432170","Cordoba")
    proyecto1=ProyectoDeLey("Etiquetados","la ley busca...")
    proyecto2=ProyectoDeLey("Reduccion de impuestos","la ley busca...")
    diputado1.agregar_proyecto(proyecto1)
    proyecto1.agregar_legislador(diputado1)
    diputado2.agregar_proyecto(proyecto1)
    proyecto1.agregar_legislador(diputado2)
    senador1.agregar_proyecto(proyecto1)
    senador1.presentar_proyecto(proyecto1,datetime(2024,2,17,18))
    asesor1=Asesor("Juan Cornalis",29,"40132976","Especialidad x")
    senador1.agregar_asesores(asesor1)
    asesor1.agregar_legislador(senador1)
    C1=Congreso()
    C1.agregar_persona(diputado1)
    C1.agregar_persona(diputado2)
    C1.agregar_persona(senador1)
    C1.agregar_persona(senador2)
    C1.ver_presentaciones()
    
```
